RecursiveCharacterTextSplitter is an advanced text splitter in LangChain that recursively tries multiple separators (such as paragraphs, lines, spaces, and finally individual characters) until it creates chunks that satisfy the specified chunk_size. It preserves the structure and meaning of the text better than CharacterTextSplitter, making it the most commonly used splitter for RAG (Retrieval-Augmented Generation) applications.

```python
# take this entire pdf and convert to text
# how to recursively split the text by characters
# take all the documents and split them into chunks of 500 characters each
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents=text_splitter.split_documents(docs)   
```

CharacterTextSplitter is a basic text splitter in LangChain that divides a document into smaller chunks using a single specified separator, such as paragraphs (\n\n), new lines (\n), spaces ( ), or commas. It attempts to split the text only where that separator appears, making it simple and efficient. However, if the separator is not found within the specified chunk_size, it may produce chunks larger than the desired size.

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter=CharacterTextSplitter(separator="\n\n",chunk_size=100,chunk_overlap=20)
text_splitter.split_documents(doc_1)
```

HTMLHeaderTextSplitter is a specialized text splitter designed for HTML documents and web pages. Instead of splitting based on characters, it splits the content according to HTML header tags like <h1>, <h2>, and <h3>. This keeps each heading together with its corresponding content, preserving the logical structure of the webpage and improving retrieval accuracy in RAG systems.

```python
import requests
from langchain_text_splitters import HTMLHeaderTextSplitter

# URL
url = "https://www.geeksforgeeks.org/machine-learning/embeddings-in-machine-learning"

# Download the HTML
html_string = requests.get(url).text

# Define which HTML headers to split on
headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
]

# Create the splitter
html_splitter = HTMLHeaderTextSplitter(
    headers_to_split_on=headers_to_split_on
)

# Split the HTML
docs = html_splitter.split_text(html_string)

# Display page content and metadata
for i, doc in enumerate(docs, start=1):
    print(f"\nChunk {i}")
    print("-" * 60)
    print("Page Content:")
    print(doc.page_content)

    print("\nMetadata:")
    print(doc.metadata)

    print("=" * 80)
```
