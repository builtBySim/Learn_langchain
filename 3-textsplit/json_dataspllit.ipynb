it attempts to keep nested_json ibjects  whole but wiill split tehm if needed to keep chuncks  between a min _chunk_size and max_chunk-size

if the value is  not a netsed json, but rather a very large string the string will not be split. if you need a hard cap on the chunk size consider composing this with a recursive text splitteron these chunck . There is an optional pre-proecssing step to split lists, bu first converting them to json(dict) and then splitting them as such.
